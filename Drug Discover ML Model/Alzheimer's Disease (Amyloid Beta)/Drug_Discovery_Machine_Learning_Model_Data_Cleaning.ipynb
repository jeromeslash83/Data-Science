```python
#import the dependencies
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#Target search for Amyloid beta
target = new_client.target
query_target = target.search('amyloid beta')
targets = pd.DataFrame.from_dict(query_target)
targets
```

The first entry in the search will be assigned into the selected_target variable.

```python
#The selected target will be the first entry which is the Human Amyloid Beta A4 Protein, which is seen accumulating in AD patient's brain.
selected_target = targets.target_chembl_id[0]
selected_target
```

The Bioactivity data for the Amyloid Beta A4 protein will be retrieved in this code 



```python
#The Bioactivity data will be retrieved.
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
df
```

The resulting data will be saved as a csv file.

```python
df.to_csv('amyloid_beta_A4_raw.csv', index=False)
```

Handling of Missing Data

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

Dropping duplicates

```python
df2_minus_dup = df2.drop_duplicates(['canonical_smiles'])
df2_minus_dup
```

Combine classes from selection variable

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_minus_dup[selection]
df3
```

Making a bioactivity class column for the data.


```python
bioactivity_class = []
for i in df3.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")

bioactivity_class
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='class')
df4 = pd.concat([df3, bioactivity_class], axis=1)
df4
```

Save the data to csv.

```python
df4.to_csv('AB_A4_raw.csv', index=False)
```
